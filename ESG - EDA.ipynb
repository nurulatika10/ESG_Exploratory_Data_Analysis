# Exploratory Data Analysis (EDA)

```python
import pandas as pd
```

```python
ESG = pd.read_csv("esg.csv")
```

```python
ESG.head()
```

```python
ESG = ESG.drop(['Symbol', 'Name', 'Address', 'Full Time Employees', 'Description', 'Industry', 'Controversy Level', 'Controversy Score', 'ESG Risk Percentile', 'ESG Risk Level'], axis=1)
```

```python
print("the unique categories of sector are")
ESG.Sector.unique()
```

```python
len(ESG.Sector.unique()) #len is for length
```

```python
print("The frequency table of sector are:\n")
ESG.Sector.value_counts()
```

```python
print("The percentage for each category of sector are:\n")
ESG.Sector.value_counts(normalize = True) *100 #normalize function is for data cleaning if needed
```

```python
ESG.Sector.value_counts().plot.bar(title = "Bar Chart of Sector in ESG Data")
```

```python
import matplotlib.pyplot as plt
```

```python
ESG.Sector.value_counts().plot.pie(autopct='%1.1f%%',
                                     title = "Pie Chart of Sector in ESG Data")

plt.axis('equal') #applied to any pie chart, and only for pie chart
plt.show()
```

```python
range_total_ESG_risk = ESG['Total ESG Risk score'].max() - ESG['Total ESG Risk score'].min()
print(f"The range of 'Total ESG Risk score' is: {range_total_ESG_risk}")
```

```python
import numpy as np
```

```python
ESG = ESG.dropna()
```

```python
np.quantile(ESG['Total ESG Risk score'], .27)
```

```python
np.percentile(ESG['Total ESG Risk score'], .27)
```

```python
#q3 - q1

q1, q3 = np.quantile((ESG['Total ESG Risk score']), [.25, .75])
manual_iqr = q3 - q1
f"The IQR of (ESG['Total ESG Risk score']) by manual compuation is {manual_iqr}"
```

```python
from scipy.stats import iqr

iqr_value = iqr(ESG['Total ESG Risk score'])
print(f"The IQR of 'Total ESG Risk score' by scipy function is {iqr_value}")
```

```python
lower_fence = np.percentile((ESG['Total ESG Risk score']), 25) - 1.5 * iqr(ESG['Total ESG Risk score'])
f"The lower fence is {lower_fence}"
```

```python
upper_fence = np.percentile((ESG['Total ESG Risk score']), 75) - 1.5 * iqr(ESG['Total ESG Risk score'])
f"The lower fence is {upper_fence}"
```
